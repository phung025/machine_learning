```python
import numpy as np
import pandas as pd
from numpy.linalg import inv
from pandas import Series, DataFrame
import seaborn as sns

%matplotlib inline
```

```python
sns.set(style="ticks", color_codes=True)
```

```python
training_data = DataFrame.from_csv('train_set1.csv', index_col=None)

# Process training data: add feature 0
cols = list(training_data.columns)
cols.insert(0,'x_0')
training_data = training_data.reindex(index=training_data.index, columns=cols, fill_value=1)

print(training_data)
```

```python
g = sns.pairplot(training_data)
```

```python
x_df = training_data[list(training_data.columns)[:-1]]
x_list = []
for row in x_df.itertuples():
    x_list.append(list(row)[1:])
X = np.array(x_list)
Y = np.array(training_data[list(training_data.columns)[-1]])
```

```python
np.matmul(np.matmul(inv(np.matmul(X.transpose(),X)), X.transpose()), Y)
```
