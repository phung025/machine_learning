```python
import pandas_datareader as pdweb
import datetime
```

```python
prices = pdweb.get_data_yahoo(['CVX', 'XOM', 'BP'], 
                              start=datetime.datetime(2010,1,1),  # Get stock data from january 1st, 2010 
                              end=datetime.datetime(2013,1,1))['Adj Close']  # Get stock data from january 1st, 2013
# ['Adj Close'] -> Closing price
prices.head()
```

```python
# The total volume of stocks traded on each day
volume = pdweb.get_data_yahoo(['CVX', 'XOM', 'BP'],
                             start=datetime.datetime(2010,1,1),
                             end=datetime.datetime(2013,1,1))['Volume']

volume.head()
```

```python
rets = prices.pct_change() # percentage change in prices
```

```python
rets.plot()
```

```python
# Correlation of the stocks
corr = rets.corr()
```

```python
prices.plot() # Plot the prices of the stocks from 2010 to 2013
# There's some correlation between the stocks
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Correlation plot
# Plot the correlation of the change in percentages of the stocks
sns.heatmap(corr)
```
